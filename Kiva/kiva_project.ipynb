# Visualizing Loans Awarded by Kiva

In this project we are visualizing insights using a dataset from <a href = "https://www.kaggle.com/fkosmowski/kivadhsv1" target = "_blank">Kaggle</a>. The dataset contains information about loans awarded by the non-profit <a href = "https://www.kiva.org/" target = "_blank">Kiva</a>. 

Using Seaborn, we are exploring the average loan amount by country using aggregated bar charts. We also visualizing the distribution of loan amounts by project type and gender using box plots and violin plots.


## Step 1: Import Necessary Python Modules


```python
from matplotlib import pyplot as plt
import pandas as pd
import seaborn as sns
```

## Step 2: Ingest The Data


```python
df = pd.read_csv('E:\\Semo4ka\\Python\\CodeCademy\\Kiva\\kiva_data.csv')
print(df.head())
```

## Step 3: Examine The Data


### Overview of the dataset:

Each entry (row) in the dataset represents a loan that Kiva awarded to a particular project. The `loan_amount` column shows the amount (in U.S. dollars) awarded to the project. The `activity` column has the category type that the project falls under. The `country` column is the country where the project is located. The `gender` column represents the gender of the primary person who applied for the loan. 


Print the first 25 rows of `df` using `.head()`

```python
print(df.head(25))
```

## Step 4: Bar Charts

Create a bar plot using Seaborn to visualize the average size of Kiva loans given to projects, by country.


```python
# Creates the figure
f, ax = plt.subplots(figsize=(15, 10))
sns.barplot(data = df, x = 'country', y = 'loan_amount')
plt.show()
```

### Adding `$` units


```python
import matplotlib.ticker as mtick

# Creates the figure
f, ax = plt.subplots(figsize=(15, 10))
# Plot the data
sns.barplot(data=df, x="country", y = "loan_amount")
fmt = '${x:,.0f}'
tick = mtick.StrMethodFormatter(fmt)
ax.yaxis.set_major_formatter(tick)
plt.show()
```

## Step 5: Learn More By Using `hue` In Your Visualization

You can visualize even more data on one bar plot by visualizing the loan amount by country, and "nesting" by gender. 

```python
# Creates the figure, you're only using this syntax so you can modify the y-axis ticks below
f, ax = plt.subplots(figsize=(15, 10))

sns.barplot(data=df, x="country", y="loan_amount", hue = 'gender')

fmt = '${x:,.0f}'
tick = mtick.StrMethodFormatter(fmt)
ax.yaxis.set_major_formatter(tick)
plt.show()
```

#### Insights from visualizations:
- On average, male recipients receive larger loans from Kiva.
- El Salvador has the least disparity in loan amounts awarded by gender among other countries.

#### Our recommendations:
- Kiva should work to decrease the gender loan gap.
- Hold workshops focused on women-led projects (to see more women-led project).
- Require that an equal amount of loans be given to male and female-driven projects.
- Review loan distribution by country.

## Step 7: Box Plots With Kiva Data

So far you have visualized the average size of loans by country using bar charts; now you are going to make a box plot to compare the distribution of loans by country.

```python
plt.figure(figsize=(16, 10))
sns.set_palette('Set3')
sns.set_context('talk')
sns.set_style('dark')
sns.boxplot(data = df, x = 'country', y = 'loan_amount')
plt.title('Distribution of loans by country')
plt.show()
```

#### Insights from visualizations:
- Kenya has the widest distribution of loans, folowed by El Salvador and Cambodia.
- The largest loan amount is most likely to be received in Cambodia, because it's average loan value is the highest, while lowes average loan value is in Philippines.

## Step 8: Box Plot by Activity


**Optional:** Set a different plot style and color palette to best visualize this data.

```python
plt.figure(figsize=(16, 10))
sns.set_palette('Accent')
sns.boxplot(data = df, x = 'activity', y = 'loan_amount')
plt.title('Distribution of loans by activity')
plt.show()
```

#### Insights from visualization:
- Farming has the widest distribution of loans and highest average loan value among other activities.
- General stores are have liest loan distribution and lowes average loan value.

## Step 9: Violin Plots

```python
plt.figure(figsize=(16, 10))

sns.violinplot(data=df, x="activity", y="loan_amount")
plt.title('Distribution of loans by activity')
plt.show()
```

### Create a violin plot that visualizes the distribution of loan amount by country.


```python
plt.figure(figsize=(16, 10))
sns.violinplot(data=df, x="country", y="loan_amount")
plt.title('Distribution of loans by country')
plt.show()
```

## Step 10: Split Violin Plots

```python
# Some styling (feel free to modify)
sns.set_palette("Spectral")
plt.figure(figsize=(18, 12))
sns.violinplot(data=df, x="country", y="loan_amount", hue = 'gender', split = True)
plt.title('Distribution of loan amounts within countries by gender')
plt.legend(bbox_to_anchor=(1.02, 1), loc='upper left', borderaxespad=0)
plt.show()
```

#### Insights from visualizations:
- In Pakistan and Cambodia, the average loan amount received by women is much less than the average loan amount received by men.
- In El Salvador, there is a fairly equal distribution of loan amounts between genders, as well as similar median loan amounts (on average).
- More male customers in Cambodia and Pakistan tend to receive higher amount of loans, while in other countries most of customers of both genders receive lower loan amounts.
